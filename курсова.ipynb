Візуалізації багатовимірних даних з кількістю ознак більше двох обов’язково необхідно за допомогою subplot збудувати більше 9 графіків за допомогою комбінування різних ознак.

```python

# Завантаження даних
dataset_path = r"C:\Users\ASUS\Downloads\data\Pokemon.csv"
pokemon_data = pd.read_csv(dataset_path)

# Вибір підмножини ознак для візуалізації (приклад)
selected_features = ['HP', 'Attack', 'Defense', 'Speed']

# Побудова subplot
fig, axes = plt.subplots(nrows=len(selected_features), ncols=len(selected_features), figsize=(15, 15))

# Заповнення subplot графіками
for i, feature1 in enumerate(selected_features):
    for j, feature2 in enumerate(selected_features):
        if i == j:
            axes[i, j].hist(pokemon_data[feature1], bins=20, color='skyblue', edgecolor='black')
            axes[i, j].set_title(feature1)
        else:
            axes[i, j].scatter(pokemon_data[feature1], pokemon_data[feature2], alpha=0.5, color='orange')
            axes[i, j].set_title(f'{feature1} vs {feature2}')

plt.tight_layout()
plt.show()
```

### Завдання №7
Вирахувати коефіцієнт пірсона для кожної пари ознак. Взяти третю ознаку за вісь X та збудувати relplot для двох ознак, що мають найбільшу кореляцію за допомогою бібліотеки seaborn.

```python

# Виберіть ознаки для обчислення кореляції
selected_features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Обчислення кореляції Пірсона
correlation_matrix = pokemon_data[selected_features].corr()

# Відображення таблиці кореляції
fig, ax = plt.subplots(figsize=(8, 6))
sns.set(font_scale=1.2)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5, ax=ax)

# Знайдіть індекс максимального значення
max_corr_index = correlation_matrix.abs().idxmax().values

# Виділіть максимальні значення жовтим кольором в таблиці
for i in range(len(max_corr_index)):
    idx = selected_features.index(max_corr_index[i])
    ax.add_patch(plt.Rectangle((idx, idx), 1, 1, fill=False, edgecolor='yellow', lw=2))

# Налаштування відображення
plt.title('Pearson Correlation Coefficients', fontsize=16)
plt.xticks(range(len(selected_features)), selected_features)
plt.yticks(range(len(selected_features)), selected_features)
plt.show()
```

```python

# Знайдіть індекси максимального значення
max_corr_indices = correlation_matrix.abs().unstack().sort_values(ascending=False).index[0]
feature1, feature2 = max_corr_indices

# Оберіть третю ознаку за вісь X
third_feature = [feature for feature in selected_features if feature != feature1 and feature != feature2][0]

# Побудова relplot для двох ознак з найбільшою кореляцією та третьої ознаки за вісь X
sns.set(style="whitegrid")
g = sns.relplot(x=third_feature, y=feature1, hue=feature2, data=pokemon_data, palette="viridis", height=6, aspect=1.2)

# Налаштування відображення
plt.title(f'Relationship Plot for {feature1} vs. {third_feature} colored by {feature2}', fontsize=16)
plt.show()
```
