### Функції та використання
Основні функції Jupyter Notebook можна розділити на такі категорії: робота з комірками, інтерфейсом, використання Markdown (проста мова розмітки, яку можна використовувати для форматування тексту).


1. **Робота з комірками**
* Створення: використовуйте кнопку “плюс” або B для нової комірки нижче, A для комірки вище.
* Виконання: Shift + Enter для виконання і переходу до наступної комірки.
* Зміна типу: M для текстової, Y для кодової.
* Очищення виведення: “Cell” у верхньому меню.
2. **Робота з інтерфейсом**
* Меню перегляду: налаштування відображення, панелі інструментів і теми оформлення.
* Вставка: використовуйте “Insert” або відповідні гарячі клавіші.
* Осередки, ядро, віджети:
* управління комірками: гарячі клавіші та опції меню для переміщення, об’єднання або поділу комірок.
* ядро (Kernel): перезапуск у разі проблем із виконанням коду.
* віджети: створення інтерактивних елементів у ноутбуці.

### Розширення
Jupyter Notebook підтримує розширення, які дають змогу додавати нові функції та можливості. Їх можна знайти в каталозі Jupyter. Перелічимо деякі з них:

- *CodeCell Formatter* – дає змогу форматувати код у комірках коду. Це може бути корисно для поліпшення читабельності коду.
- *Code Inspector* – дає змогу переглядати змінні та об’єкти в комірках коду. Це корисно для налагодження коду і розуміння його роботи.
- *Table of Contents* – додає зміст до блокнота, що допомагає в навігації блокнотом і швидкого доступу до потрібної інформації.
- *Codefolding* – дає змогу згортати і розгортати комірки коду. Це підвищує читабельність ноутбука.
- *Autopep8* – автоматично форматує код відповідно до правил автоформатування PEP8. Це може бути корисно для поліпшення читабельності коду.

import numpy as np

def play_audio():
    fs = 44100  # sample rate
    T = 5.0    # seconds
    t = np.linspace(0, T, int(T*fs), endpoint=False)  # time variable
    x = 0.5 * np.sin(2 * np.pi * 440.0 * t)  # pure sine wave at 440 Hz

    display(Audio(x, rate=fs))

Математичне сподівання (середнє значення) обчислюється за наступною формулою:

$$
\mu = \frac{1}{N} \sum_{i=1}^{N} x_i
$$
де:


* μ - математичне сподівання (середнє значення),
* N - кількість значень у вибірці,
* ∑ - символ суми,
* x_i - кожне окреме значення у вибірці.

Формула для обчислення дисперсії виглядає наступним чином:
$$
\sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2
$$

де:



* σ^2 - дисперсія,
* N - кількість значень у вибірці,
* ∑ - символ суми,
* x_i - кожне окреме значення у вибірці,
* μ - математичне сподівання (середнє значення).

````python
# Вибираємо всі стовпці числового типу
numeric_columns = df.select_dtypes(include=[np.number]).columns
df_numeric = df[numeric_columns]

# Вибираємо всі стовпці, починаючи з третього, для обчислення середнього та стандартного відхилення
data_selected = df_numeric.iloc[:, 2:]

# Обчислюємо математичне сподівання та стандартне відхилення для вибраних даних
mean_selected = data_selected.mean()
std_selected = data_selected.std()

# Дисперсія для вибраних даних
variance_selected = np.var(data_selected)

# Стандартизуємо вибрані дані
standardized_data_selected = (data_selected - mean_selected) / std_selected

# Виводимо результати
print("\nМатематичне сподівання для вибраних даних:")
print(mean_selected)
print("\nДисперсія для вибраних даних:")
print(variance_selected)
print("\nСтандартне відхилення для вибраних даних:")
print(std_selected)
print("\nСтандартизовані вибрані дані:")
print(standardized_data_selected.head(10))
````

### Завдання №7
Вирахувати коефіцієнт пірсона для кожної пари ознак. Взяти третю ознаку за вісь X та збудувати relplot для двох ознак, що мають найбільшу кореляцію за допомогою бібліотеки seaborn.

````python
# Вибір числових стовпців
numeric_columns = df.select_dtypes(include=[np.number]).columns
df_numeric = df[numeric_columns]

# Обчислення кореляції Пірсона
correlation_matrix = df_numeric.corr()

# Знаходження ознак з найбільшою кореляцією
max_corr = abs(correlation_matrix.values[np.triu_indices_from(correlation_matrix, 1)]).max()
idx = np.where(abs(correlation_matrix) == max_corr)
max_corr_features = [correlation_matrix.columns[idx[0][0]], correlation_matrix.columns[idx[1][0]]]

# Виведення таблиці кореляції Пірсона
print('Таблиця кореляції Пірсона:')
print(correlation_matrix)

# Знайдіть індекси максимального значення
max_corr_indices = correlation_matrix.abs().unstack().sort_values(ascending=False).index[0]
feature1, feature2 = max_corr_indices

# Оберіть третю ознаку за вісь X
x_feature = [feature for feature in numeric_columns if feature != feature1 and feature != feature2][0]

# Побудова relplot для двох ознак з найбільшою кореляцією та третьої ознаки за вісь X
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))

# Побудова relplot
sns.relplot(x=x_feature, y=max_corr_features[0], data=df, kind="scatter", height=8, aspect=1.2)
plt.title(f'Relplot для {max_corr_features[0]} та {max_corr_features[1]}', fontsize=16)
plt.xlabel(x_feature, fontsize=14)
plt.ylabel(max_corr_features[0], fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
````

### Завдання №9
Видалити також випадкові значення з ознак що мають найбільшу кореляцію у кількості 10% для обох. Підрахувати кореляцію. Збудувати multipple scatterplot за допомогою FacetGrid для цих оновлених ознак використовуючи бібліотеку seaborn взявши третю ознаку за вісь x.

````python


correlation_matrix = pokemon_data.corr(method="pearson", numeric_only=True)


correlated_features = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)

# Remove 10% of random values for selected features
for feature in correlated_features:
    mask = np.random.rand(len(pokemon_data)) < 0.9
    pokemon_data.loc[~mask, feature] = np.nan


print("Correlation after removing random values:")
print(pokemon_data.corr(method="pearson", numeric_only=True))
print("There are not significant changes")

# Build multiple scatterplots using FacetGrid
g = sns.FacetGrid(pokemon_data, col='Legendary', hue='Legendary', height=4, legend_out=True)
g.map(sns.scatterplot, 'Attack', 'Defense', 'Speed')
g.add_legend()
plt.show()
````
